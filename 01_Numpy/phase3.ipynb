```python
%pip install -q matplotlib numpy
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Data Structure: [restaurant_id, 2021, 2022, 2023, 2024]


sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato Sales Data ====")

print("\n sales data shape",sales_data.shape)

print("\n sales of first 3 data of restaurant\n",sales_data[:3])
```

```python
# total sales / year

total_sales = np.sum(sales_data[:, 1:], axis=0)
print("\n Total sales per year:\n", total_sales)
print("\n yearly sum:\n", np.sum(total_sales))


#Minimum and Maximum Sales
min_sales = print("min sale: ",np.min(sales_data[:, 1:], axis=1))
max_sales = print("sales per year:",np.max(sales_data[:, 1:], axis=0))

# Average Sales
average_sales = np.mean(sales_data[:, 1:], axis=1) 
print("\n Average sales per year:\n", average_sales)

# Plotting the sales data
years = np.array([2021, 2022, 2023, 2024])
plt.figure(figsize=(10, 6))
for i in range(sales_data.shape[0]):
    plt.plot(years, sales_data[i, 1:], marker='o', label=f'Restaurant {int(sales_data[i, 0])}')
```

```python
#cumalative sum of sales data

cumsum = np.cumsum(sales_data[:, 1:], axis=1)
cumsum
#plt.title('Sales Data of Restaurants Over Years')


# Plotting the sales data

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title('Cumulative Sales Data of Restaurants Over Years')
plt.xlabel('Year'), plt.ylabel('Cumulative Sales')
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([5, 4, 3, 2, 1])

# Vector Operations
sum= vector1 + vector2
print("\nSum of vectors:", sum) 

multiplication = vector1 * vector2
print("Element-wise multiplication of vectors:", multiplication)

dot_product = np.dot(vector1, vector2)
print("Dot product of vectors:", dot_product)

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
angle_degrees = np.degrees(angle)
```
